# LinearRegression From Scratch using just loops
 

#### Implementing just for loops // comparing the results with sklearn LinearRegression model

#### Importing Libraries 

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LinearRegression
```

#### Data Initialization

```python
X = np.array([1, 2, 3, 4, 5, 6])
y = np.array([5, 7, 9, 11, 13, 15])
```

#### Parameter Initialization

```python
w = 0.0 ## Default kept to zero
b = 0.0 ## Default kept to zero
learning_rate = 0.01    ### changed from 0.001 to 0.01 after hit and trial 
epochs = 10000
n = len(y)
```

#### Linear Regression Model

```python
for i in range(epochs):
    loss = 0
    db = 0                  ### Loop for training cycle
    dw = 0

    for j in range(n):
        y_pred = w * X[j] + b

        loss = loss + (y_pred - y[j]) ** 2          ### Implementing loop for total avg loss 

    total_loss = loss / (2 * n)


    for j in range(n):
        d_w = ((w * X[j] + b) - y[j]) * X[j]
        d_b = ((w * X[j] + b) - y[j])

        dw = dw + d_w                           ### Implementing loop for Gradient Updation
        db = db + d_b

    w = w - learning_rate * (dw/n)
    b = b - learning_rate * (db/n)


    if i % 100 == 0:
        print(f"Epoch {i}: Loss = {float(total_loss):.4f}, w = {float(w):.4f}, b = {float(b):.4f}")


print(f"\nFinal parameters -> w = {float(w):.3f}, b = {float(b):.3f}")

y_pred = w * X + b                              ### calculating y_pred for all the values of x

ss_res = ((y - y_pred) ** 2).sum()               # residual sum of squares
ss_tot = ((y - y.mean()) ** 2).sum()             # total sum of squares
r2 = 1 - (ss_res / ss_tot)
print(f"R^2 Score: {r2:.4f}")               ## Gives perfect output as 1
```

```python
plt.scatter(X, y, color='Blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Fitted line')
plt.xlabel('TV Advertisement Budget')                        ### Plotting for visiual understanding
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
lig = LinearRegression()

lig.fit(X.reshape(-1,1),y)

y_pred = lig.predict(X.reshape(-1,1))            ### Gives the same perfect score of 1 

print('r2 score = ',lig.score(X.reshape(-1,1), y))

```
